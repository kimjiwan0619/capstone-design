```python
# 텍스트 파일을 읽어서 리스트로 저장합니다.
with open('/data/kimjiwan0619/capstone-design/yolov8_tracking/runs/track/exp/tracks/video1.txt', 'r') as f:
    lines = f.readlines()

# 각 줄을 분할하여 마지막 정보를 추출합니다.
data = []
for line in lines:
    values = line.strip().split()
    last_info = values[1]
    data.append((line, last_info))

# 마지막 정보를 기준으로 오름차순으로 정렬합니다.
sorted_data = sorted(data, key=lambda x: x[1])

# 정렬된 결과를 출력합니다.
for item in sorted_data:
    print(item[0])
```
